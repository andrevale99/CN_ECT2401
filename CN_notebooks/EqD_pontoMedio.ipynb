```python
import numpy as np

def df(x, y):
  return (10-y)/(1-x**2)

def pontoMedio(a, b, h, y0):
  x = np.arange(a, b+0.1, h)
  print(x)
  y = np.zeros((1, len(x)))

  y[0][0] = y0

  for i in range(0, len(x)-1):
    k1 = df(x[i], y[0][i])
    k2 = df(x[i] + h/2, y[0][i]+k1*h/2)

    y[0][i+1] = y[0][i] + k2*h

  print(y)



pontoMedio(2., 2.8, 0.4, 2.4)
```
