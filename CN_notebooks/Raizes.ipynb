```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *

def f(x):
    return x**2/5 - 2*x -3
```

```python
def Bissec(a, b, erro):
  x = a
  x_old = x

  erro_final = 1.0001

  while erro_final > erro:
    x_old = x

    x = (a+b)/2

    if f(a)*f(x) < 0:
      b = x
    else:
      a = x

    erro_final = np.abs((x-x_old)/x)
  
  print(f'{a}\t{b}\t{x}')

#==============================================================================

def Falsa_pos(a, b, erro):
  x = a
  x_old = x

  erro_final = 1.0001

  while erro_final > erro:
    x_old = x

    x = a - ((a-b)*f(a))/(f(b)-f(a))

    if f(a)*f(x) < 0:
      b = x

    else:
      a = x

    erro_final = np.abs((x-x_old)/x)

  print(f'{a}\t{b}\t{x}')

  #==============================================================================
def Mtd_Newtow(x0, erro):
  x_n1 = 0
  x_n = x0
  erro_final = 1.0001

  x = symbols('x')

  derivada = diff(func,x)

  while True:
    #x_n1 = x_n - func(x_n)/((f(x_n+0.0001)-f(x_n)))
    x_n1 = x_n - func(x_n) / derivada(x_n)
    erro_final = np.abs((x_n1-x_n)/x_n1)

    if erro_final < erro:
      break

    x_n = x_n1

  print(f'{x_n}\t{x_n1}\terro:{erro_final}')
```

```python
Bissec(10., 12., 0.001)
Falsa_pos(10., 12., 0.001)

x = symbols('x')
funcao = (x**2)/5 - 2*x - 3
Mtd_Newtow(11.4, 1e-3, funcao)
```
