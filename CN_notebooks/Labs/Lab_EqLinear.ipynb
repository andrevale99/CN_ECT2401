```
 def decomposicaoLU(A):
  U = A

  dim = list(np.shape(U))
  #print(dim)

  lista_fatores = []

  #print(f'\n0:\n {A}')

  for c in range(0, dim[0]-1):

    for l in range(c+1, dim[0]):
      pivo = U[c][c]
      fator = U[l][c]/pivo
      lista_fatores.append(fator)
      U[l][:] = U[l][:] - fator*U[c][:]

    #print(f'{c+1}:\n {U}\n')

  #print(f'Fatores: {lista_fatores}')

  #print(f'Matriz U:\n{U}',"\n")

  L = np.identity(dim[0])
  
  aux = 0
  for i in range(1, dim[0]):
    for j in range(0, i):
      L[i][j] = lista_fatores[aux]
      aux+=1

  #print(L)

  return L, U
```

```
L, U = decomposicaoLU(A)

print("Matriz L:\n",L,"\n\n","Matriz U:\n",U)
```
