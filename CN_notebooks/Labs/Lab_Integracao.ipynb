```python
import numpy as np

#==============================================================================

def f(x):
  return (5 * np.sqrt(x))/x
 
#==============================================================================

def regraTrapezio(a,b,n):
  h = (b-a)/n

  x = np.arange(a,b+0.1, h)
  y = f(x)

  A = 0

  dim = np.shape(y)

  for i in range(0, dim[0]):
    if (i==0) or (i==dim[0]-1):
      A += y[i]
    else:
      A += 2*y[i]

  A = (h/2)*A
  print(f'Resultado pelo Mtd. do Trapezio: {A}')

  return A

#==============================================================================

def simpson13(a, b, n):
  h = (b-a) / (2*n)

  A = 0

  x = np.arange(a,b+0.1, h)
  y = f(x)

  A = y[0] + y[len(y)-1]

  for i in range(1, len(y)-1):
    if i%2==0:
      A += 2*y[i]
      #print(2*y[i])

    else:
      A += 4*y[i]
      #print(4*y[i])

  A = h/3*A

  print(f'Resultado pelo Simpson 1/3: {A}')

  return A

#==============================================================================

def simpson38(a,b, n):

  h = (b-a)/(3*n)

  x = np.arange(a, b+0.1, h)

  y = f(x)

  A = 0
  A = y[0] + y[len(y)-1]
 

  for i in range(1, len(y)-1):
    if i%3==0:
      A += 2*y[i]
    else:
      A += 3*y[i]

  A = (3*h)/8 * A
  
  print(f'Resultado pelo Simpson 3/8: {A}')
  
  return A
```

```python
n = 6
a = 1.2
b = 3.5
A_trp = regraTrapezio(a,b,n)

n = 6
a = 1.2
b = 3.5
A_s13 = simpson13(a,b,n)

n = 6
a = 1.2
b = 3.5
A_38 = simpson38(a,b,n)
```
