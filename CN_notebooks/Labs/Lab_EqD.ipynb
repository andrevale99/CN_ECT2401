```python
import numpy as np

def df(x, y):
  return (1+x)*np.sqrt(y)

def euler(a, b, h, y0):
  x = np.arange(a, b+0.1, h)
 # print(x)
  
  y = np.zeros((1, len(x))).reshape(len(x))
  y[0] = y0

  for i in range(0, len(x)-1):
    y[i+1] = y[i] + h * df(x[i], y[i])

  #print(y)
  return y
#================================================================
def heun(a, b, h, y0):
  x = np.arange(a, b+0.1, h)
  #print(x)
  y = np.zeros((1, len(x))).reshape(len(x))
  
  y[0] = y0

  for i in range(0, len(x)-1):
    k1 = df(x[i], y[i])
    k2 = df(x[i]+h, y[i] + k1*h)
    k = (k1+k2)/2

    y[i+1] = y[i] + k*h

  #print(y)
  return y
#================================================================
def pontoMedio(a, b, h, y0):
  x = np.arange(a, b+0.1, h)
  #print(x)
  y = np.zeros((1, len(x))).reshape(len(x))

  y[0] = y0

  for i in range(0, len(x)-1):
    k1 = df(x[i], y[i])
    k2 = df(x[i] + h/2, y[i]+k1*h/2)

    y[i+1] = y[i] + k2*h

  #print(y)
  return y
#================================================================
def rk4(a, b, h, y0):
  x = np.arange(a, b+0.1, h)
  
  y = np.zeros((1, len(x))).reshape(len(x))
  
  y[0] = y0

  for i in range(0, len(x)-1):
    k1 = df(x[i], y[i])
    k2 = df(x[i]+h/2, y[i]+k1*h/2)
    k3 = df(x[i]+h/2, y[i]+k2*h/2)
    k4 = df(x[i]+h, y[i]+k3*h)

    k = (k1 + 2*k2 + 2*k3 + k4)/6

    y[i+1] = y[i] + h*k

  #print(y)
  return y
```

#**Q3**

```python
a = 2.2
b = 4.3
h = 0.3
y0 = 2.5

E = euler(a, b, h, y0)
H = heun(a, b, h ,y0)
RK = rk4(a, b, h, y0)
```

```python
import matplotlib.pyplot as plt

x = np.arange(a, b+0.1, h)

plt.scatter(x, E, label="euler")
plt.scatter(x, H, label="heun")
plt.scatter(x, RK, label="rk4")

plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

plt.show()
```
