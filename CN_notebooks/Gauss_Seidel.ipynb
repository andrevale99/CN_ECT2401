```
matriz = np.array([[0 , -1/3, 1/3],
                  [-0.4, 0, -0.1],
                  [-0.25, 0.25, 0]], dtype='float16')

x = np.array([[1],
              [1],
              [1]], dtype='float16')

d = np.array([[2],
              [2.3],
              [-1.25]], dtype='float16')

x_old = x

print(x)
```

```
x[0][0] = matriz[0][0]*x[0][0] + matriz[0][1]*x[1][0] + matriz[0][2]*x[2][0] + d[0]
print(x, "\n")

x[1][0] = matriz[1][0]*x[0][0] + matriz[1][1]*x[1][0] + matriz[1][2]*x[2][0] + d[1]
print(x, "\n")

x[2][0] = matriz[2][0]*x[0][0] + matriz[2][1]*x[1][0] + matriz[2][2]*x[2][0] + d[2]
print(x, "\n")

er = np.zeros((1,1), dtype='float32')

er = max(np.abs(x-x_old))/max(np.abs(x))

print(f'Erro relativo: {er}')

```
