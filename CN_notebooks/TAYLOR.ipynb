```
import matplotlib.pyplot as plt
import numpy as np
from sympy import *

# Serve para achar a esxpressão da derivada em python
#para colocar na funçaõ
x = symbols('x')
funcao = 5*x**3 + x +2

vetor = [funcao]

for i in range(0,3):
  derivada = diff(funcao)
  vetor.append(derivada)  
  
  funcao = derivada
  
print(vetor)
del vetor

del x

def f(x):
  return 5*x**3 + x + 2

def P2(x):
  return 15*x**2 - 14*x +7

def R2(x):
  return 30 * ((x-x0)**3)/6

def colocar_val_fx(x,y):
  for i in range(0, len(x)):
    y[i] = f(x[i])


x0 = 1
x = 1.1

#x0 = float(input("Digite o centro(x0):"))
#x = float(input("Digite o valor que deseja analisar (x): "))

print("valor de F(x): ", f(x))
print("Valor de P2(x): ", P2(x))
print("Valor do R2(t): ", R2(1.1))

eixo_x = np.arange(-5, 5, 0.05)
y_f = np.zeros((len(eixo_x)))

colocar_val_fx(eixo_x, y_f)

poli_2 = P2(eixo_x)

#for i in range(0, len(poli_2)):
 #   poli_2[i] = P2(x, eixo_x[i])


plt.plot(eixo_x, y_f, lw='1.5', color='black', ls='-', label='fx')
plt. plot(eixo_x, poli_2, lw='1.', color='red', ls='--', label='P2')

plt.grid()
plt.xlabel("X")
plt.ylabel("e^(x/2)")
plt.xlim(-2, 2)
plt.ylim(-2, 5)
plt.title("Análise")
plt.legend()

plt.show()
```

```

def P2(x):
  return np.log(2.5) + (1/2.5)*(x-2.5) - (1/(2*(2.5)**2))*(x-2.5)**2

print(P2(3))
```

```
xg = np.arange(1,4,0.05)
yg = f(xg)
ygp = P2(xg)

plt.plot(xg,yg,'r')
plt.plot(xg,ygp,'b')
plt.show()
```
