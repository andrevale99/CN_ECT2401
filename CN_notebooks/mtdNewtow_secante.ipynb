```python
#   Bibliotecas que poderão ser usadas 
#para as lógicas

import matplotlib.pyplot as plt
import numpy as np
from sympy import *
from random import random

def f(x):
    return np.exp(-x/4) - 0.5

def deri_f(x):
    return -1/4*np.exp(-x/4)
```

```python
#Metodo de Newtow

quant_itr = 0

x_n1 = 0

er = 0

x_n = 0.323

while True:
    quant_itr += 1

    x_n1 = x_n - f(x_n)/deri_f(x_n)

    er = np.abs((x_n1-x_n)/x_n1)

    print("Erro: ", er)
    print("Valor da função: ", x_n1)
    print("Quantidade de iterações: ", quant_itr)
    print()

    if er < 1e-3:
        break

    x_n = x_n1

del quant_itr
del x_n
del er

```

```python
x = np.arange(x_n1-0.1, x_n1+0.1, 0.05)
y = f(x)

ponto = x_n1

plt.scatter(x_n1, 0, color='red', marker='o', label='Aprox')
plt.plot(x ,y, color='black', lw=1.5, label='Função')

plt.title("Aproximação")
plt.grid()
plt.xlabel("X")
plt.ylabel("f(x)")
plt.legend()

plt.show()
```
