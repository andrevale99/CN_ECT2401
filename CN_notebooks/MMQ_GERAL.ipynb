```
import matplotlib.pyplot as plt
import numpy as np

def f0(x):
  return x

def f1(x):
  return x**3

x = np.array([0.9, 1.4, 1.9])
y = np.array([3.3, 2.7, 2.5])

```

```
mf = np.zeros((2,2))

for i in range(0, 2):
  for j in range(0, len(x)):
    if i == 0:
      mf[0][i] += f0(x[j])*f0(x[j])
    if i == 1:
      mf[0][i] += f1(x[j])*f0(x[j])

for i in range(0, 2):
  for j in range(0, len(x)):
    if i == 0:
      mf[1][i] += f1(x[j])*f0(x[j])
    if i == 1:
      mf[1][i] += f1(x[j])*f1(x[j])

soma = 0  
for i in range(0,len(x)):
  soma += f0(x[i])*y[i]

soma2 = 0  
for i in range(0,len(x)):
  soma2 += f1(x[i])*y[i]

my = np.array([
               [soma],
               [soma2]
])

print(my,"\n")

sols = np.linalg.inv(mf) @ my

print(mf,"\n")
print(sols)
```
