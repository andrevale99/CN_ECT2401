# **Regra do Simpson 1/3 SIMPLES** 


A integração utilizando a regra de Simpson 1/3, é um método de integração que aproxima a função por um polinômio de segundo grau.

Primeiro devemos ter três pontos para achar o polinômio de segundo grau. Dois deles são as extremidades (o limite superior $b$ e inferior $a$) e o terceiro é o ponto médio entre o intervalo [$a$,$b$] (para facilitar os cáculos.

Após achar os três pontos, devemos definir o intervalo $h$ que é:

$h = \frac{b-a}{2}$

Depois podemos achar o tamanho dos intervalos pela equação:

$x_i = x_0 + ih$

Onde $x_0$ representa o ponto inicial, $x_i$ representa o ponto qual deseja saber e o $i$ qual intervalo deseja saber ($i=1,2,3,...,n$).

Para achar o polinômio, devemos utilizar um dos métodos de Interpolação. Aqui usaremos o de Lagrange.

A interpolação de Lagrange é dada pela equação:

$P_n(x) = \sum\limits_{i=0}^{n}y_iL_i(x)$

```python
import numpy as np

def f(x):
  return np.sqrt(1+(10*x-10)**2)

n = 1

a = 1
b = 5

h = (b-a) / (2*n)

print(h)
```

```python
x = np.arange(a,b+0.1, h)

x[0] = a

for i in range(1, len(x)):
  x[i] = h + x[i-1]

print(f'Intervalo de X:\n{x}\n')

y = f(x)
print(f'Intervalo de Y:\n{y}\n')
```

```python
def Simpson_13(x, y, h):
  A = 0

  A = y[0] + y[len(y)-1]

  for i in range(1, len(y)-1):
    if i%2==0:
      A += 2*y[i]
      #print(2*y[i])

    else:
      A += 4*y[i]
      #print(4*y[i])

  A = h/3*A

  print(f'Resultado aproximado da integral: {A}')
```

```python
Simpson_13(x,y, h)
```

```python
import numpy as np

def Simpson_13(x, y, h):
  A = 0

  A = y[0] + y[len(y)-1]

  for i in range(1, len(y)-1):
    if i%2==0:
      A += 2*y[i]
      #print(2*y[i])

    else:
      A += 4*y[i]
      #print(4*y[i])

  A = h/3*A

  print(f'Resultado aproximado da integral: {A}')

def f(x):
  return np.exp(x) - np.exp(x/2)

n = 2

a = 1.2
b = 3.9

h = (b-a) / (2*n)

print(h)

x = np.arange(a,b+0.1, h)

x[0] = a

for i in range(1, len(x)):
  x[i] = h + x[i-1]

print(f'Intervalo de X:\n{x}\n')

y = f(x)
print(f'Intervalo de Y:\n{y}\n')

Simpson_13(x,y, h)
```
