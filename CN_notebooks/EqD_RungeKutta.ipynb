```python
import numpy as np

def df(x, y):
  return -0.09*np.sqrt(np.abs(y))

def RungeKutta(a, b, h, y0):
  x = np.arange(a, b+0.1, h)
  print(x)

  y = np.zeros((1, len(x))).reshape(len(x))
  
  y[0] = y0

  for i in range(0, len(x)-1):
    k1 = df(x[i], y[i])
    k2 = df(x[i]+h/2, y[i]+k1*h/2)
    k3 = df(x[i]+h/2, y[i]+k2*h/2)
    k4 = df(x[i]+h, y[i]+k3*h)

    k = (k1 + 2*k2 + 2*k3 + k4)/6

    y[i+1] = y[i] + h*k

  print(y)

RungeKutta(0., 3.6375, 0.7275, 6.8)
```

```python
import matplotlib.pyplot as plt

x = np.arange(6.8, 0., -0.05)

plt.plot(x, df(x, x))

plt.show()
```
