# **INTEGRAÇÃO DE SIMPSON 3/8 COMPOSTA**

Como na regra de simpson 1/3, dado um conjunto de $n$ pontos ($x$, $y$), iremos subdividir no intervalo [$a$,$b$] em $n$ intervalos, onde cada conjunto de 4 pontos (3 subintervalos), iremos realizar o método de simpson 3/8 e somar sua áreas para encontrar o resulatdo da integral.

Uma das equação que muda é dos intervalos $h$ o qual ficará:

$h = \frac{b-a}{3n}$

Lembrando que o denominador $3n$ deve ser a quantidade de subintervalos, caso a questão queira com 3 aplicações, o $n=3$, porém, se a questão quer 9 subintervalos, o valor de $n$ deve ser 3 para que a multiplicação $3n = 9$.
**Só podemos utilizar esse método de integração se os subintervalos for múltiplo de 3**

```python
import numpy as np

def Simpson38(y, h):
  A = 0
  A = y[0] + y[len(y)-1]
 

  for i in range(1, len(y)-1):
    if i%3==0:
      A += 2*y[i]
    else:
      A += 3*y[i]

  A = (3*h)/8 * A
  A = round(A, 4)
  print(f'Resultado da integração 3/8: {A}')

def f(x):
  return (x**2)/5 + 2*np.sin(x)

a = -2
b = 2
n = 2

h = (b-a)/(3*n)
print(f'Intervalos h de: {h}\n')

x = np.arange(a, b+0.1, h)
x = np.round(x, 4)
print(f'x_i\n:\n{x}\n')

y = f(x)
y = np.round(y, 4)
print(f'f(x_i)\n:{y}\n')

Simpson38(y, h)
```
