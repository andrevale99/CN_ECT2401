```python
import numpy as np

def df(x,y):
  return -0.19*np.sqrt(np.abs(y))

def euler(a, b, h, y0):
  x = np.arange(a, b+0.1, h)
  print(x)
  
  y = np.zeros((1, len(x)))
  y[0][0] = y0

  for i in range(0, len(x)-1):
    y[0][i+1] = y[0][i] + h * df(x[i], y[0][i])

  print(y)
 

euler(, 4.18, -0.445, 6.8)
```
