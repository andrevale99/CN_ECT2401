# **INTERPOLAÇÃO DE LAGRANGE**

A equação mais simples que pode representar qualquer outra equação sãp ps polinômios.

Dado um conjunto de ponto no palno existe somente um polinômio que passe por todos os pontos.

Para encontrar esse polimômio **Interpolação de Lagrange** que consiste em criar um polinômio com os dados em mão.

Dado um conjunto de pontos:

(-1; 3),(2; 6), (4; -2)

Para saber o grau do polinômio basta pegar a quantidade de pontos e contar começando do 0:

$(x_i; y_i)\ para\  i=0, 1, 2, ..., n$

Sendo assim para nosso estudo, como temos 3 pontos, o polinômio será de segundo grau.

Para encontrar a equação, iremos utilizar de uma equação que corresponde a cada ponto.

image.png

A equação é formada a partir das abcissas do ponto $x_i$. Exemplo:

$L_0(x) = \frac{(x-2)(x-4)}{(-1-2)(-1-4)}$

$L_1(x) = \frac{(x-4)(x-(-1))}{(2-(-1))(2-4)}$

$L_2(x) = \frac{(x-2)(x-(-1))}{(4-(-1)(4-2)}$

Na parte do **numerador** utilizamos a multiplicação quadrática dos pontos sem utilizar a do corresponde. Exemplo: Quando fomos calcular o $L_0(x)$ não utilizamos o $x_0 = -1$, somente o restante.

Já na parte do **denominador** utilizamos o $x_0$, mas isso serviu para deixar divido por 1, ou seja, sobra somente a multiplicação quadrática de cima (vamos ver o restante do cálculo)

Após achar os termos $L_n(x)$ basta somarmos as equações e multiplicar seu respectivo $L_n(x)$ com seu $y_i$ do ponto. Resumindo:

$P_n(x) = \sum\limits_{i=0}^{n}y_iL_i(x)$

Para os pontos dados, a equação se torna:

$P2(x) = y_0L_0(x) + y_1L_1(x) + y_2L_2(x)$

$P2(x) = 3 L_0(x) + 6 L_1(x) - 2 L_2(x)$



Por fim basta substituirmos os termos de $L_i$ na equação a cima e encontrar o polinômio. Não precisamos desenvolver a equação, pois o resultado é o mesmo.

# **EXEMPLO**

Bibliotecas para utilizarmos para as resoluções e visualizações dos dados

```
import matplotlib.pyplot as plt
import numpy as np
```

A função a baixo calcula o somatório do polinômio $P_n(x)$ e ao mesmo tempo resolve o polinômio no ponto $a$

```

def InterLgr(x,y, a):
  dim = np.shape(x)
  S = 0
  for i in range(0, dim[0]):
    L = 1
    for j in range(0, dim[0]):
      if i != j:
        L = L * (a-x[j])/(x[i]-x[j])

    S = S + y[i]*L

  return S
```

```
x = np.array([ 29,38, 41])
y = np.array([np.log10(29), np.log10(38), np.log10(41)])
```

```
t = InterLgr(x,y,35)
print(t)
```

```
print(np.log10(35))
print(t - np.log10(35))
```

Por fim mostra que o polinômio encontrado passa pelos pontos.

```
xgrafico = np.arange(-6,8,0.05)
ygrafico = InterLgr(x,y,xgrafico)

plt.scatter(x,y)
plt.plot(xgrafico,ygrafico)
plt.grid()

plt.show()
```
