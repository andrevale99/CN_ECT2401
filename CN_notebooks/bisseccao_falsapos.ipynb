```python
#   Bibliotecas que poderão ser usadas 
#para as lógicas

import matplotlib.pyplot as plt
import numpy as np
from sympy import *

def f(x):
    return np.log(x) - 0.9
```

```python
#Metodo da Bissecção 

quant_it = 0

a = 1.92
b = 3.78

x = a

while quant_it < 3:
    x_old = x

    x = (a+b)/2

    if (f(a)*f(x)) < 0:
        b = x
    else:
        a = x

    er = np.abs((x-x_old)/x)

    #if er < 10**-2:
        #break
    quant_it += 1

print(f'{er}\t |{a} {b}| {x}')


```

```python
#Metodo da Falsa Posição

a = -1
b = 2

x = a

while 1:
    x_old = x

    x = a - (((b-a)*f(a))/(f(b)-f(a)))

    if (f(x)*f(a)) < 0 :
        b = x
    else:
        a = x

    er = np.abs(x-x_old)/np.abs(x)

    if(er < 0.01): 
        
        break

print(f'{er}\t|{a} {b}| {x}')
```
